```python
import pandas as pd

df=pd.read_csv('/content/Toyota.csv')
df
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
# Remove duplicate observations
df.drop_duplicates(inplace=True)
```

```python
# Check for and fix structural errors, such as inconsistent data types or formatting issues
# Convert 'Doors' column to numeric (assuming it should be a numeric feature)
df['Doors'] = pd.to_numeric(df['Doors'], errors='coerce')  # coerce will convert non-numeric values to NaN

# Convert 'FuelType' to categorical data type
df['FuelType'] = df['FuelType'].astype('category')
```

```python
df.head()
```

```python
# Define a function to identify and filter outliers using the interquartile range (IQR) method
def remove_outliers(column):
    if pd.api.types.is_numeric_dtype(column):
        Q1 = column.quantile(0.25)
        Q3 = column.quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        return column[(column >= lower_bound) & (column <= upper_bound)]
    else:
        return column

# Apply the function to each column containing potential outliers
for col in df.columns:
    df[col] = remove_outliers(df[col])

df
```

```python
# Check for missing data
missing_data = df.isnull().sum()
print("Missing data:\n", missing_data)

# Handle missing data based on the specific context of your dataset
# For numeric columns, replace missing values with mean
numeric_columns = df.select_dtypes(include='number').columns
for col in numeric_columns:
    df[col].fillna(df[col].mean(), inplace=True)  # Replace missing values with mean

# For categorical columns, you can replace missing values with the most frequent category
df['MetColor'].fillna(df['MetColor'].mode()[0], inplace=True)  # Replace missing values with mode
df['FuelType'].fillna(df['FuelType'].mode()[0], inplace=True)  # Replace missing values with mode

df
```

```python
# Standardize text data in categorical columns
df['FuelType'] = df['FuelType'].str.strip().str.lower()
```

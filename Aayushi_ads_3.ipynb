```python
import pandas as pd
df = pd.read_csv('Crime1.csv')
df
```

# **Matplotlib Plots:**

**Bar Plot of Category Counts:**

This plot shows the distribution of incident categories in the dataset.
Each bar represents a category, and the height of the bar represents the number of incidents belonging to that category.
It helps in understanding which categories of incidents are more frequent than others.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming 'df' is your DataFrame containing the dataset
category_counts = df['Category'].value_counts()
plt.figure(figsize=(12, 6))
category_counts.plot(kind='bar', color='skyblue')
plt.title('Count of Incidents by Category')
plt.xlabel('Category')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

**Scatter Plot of Coordinates (X, Y):**

This plot visualizes the spatial distribution of incidents based on their coordinates (X, Y).
Each point on the plot represents an incident, and its position corresponds to its X and Y coordinates.
It helps in identifying spatial patterns or clusters of incidents, which might indicate areas with higher or lower incident rates.

```python
plt.figure(figsize=(10, 8))
plt.scatter(df['X'], df['Y'], alpha=0.5, color='orange')
plt.title('Scatter Plot of Incident Locations')
plt.xlabel('X')
plt.ylabel('Y')
plt.grid(True)
plt.show()
```

# **Seaborn Plots**

**Count Plot of DayOfWeek:**

This plot shows the distribution of incidents across different days of the week.
Each bar represents a day of the week, and the height of the bar represents the count of incidents that occurred on that day.
It helps in understanding whether there are any patterns or trends in incident occurrence based on the day of the week.

```python
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.countplot(x='DayOfWeek', data=df, palette='Set2')
plt.title('Count of Incidents by Day of Week')
plt.xlabel('Day of Week')
plt.ylabel('Count')
plt.show()
```

**Violin Plot of Category and X:**

This plot visualizes the distribution of incident locations (X-coordinate) across different categories.
Each violin plot represents a category, and the width of the plot at a particular X-coordinate indicates the density of incidents in that category at that location.
It helps in identifying whether there are any differences in the spatial distribution of incidents across different categories.

```python
plt.figure(figsize=(12, 6))
sns.violinplot(x='Category', y='X', data=df, palette='pastel')
plt.title('Violin Plot of Incident Locations by Category')
plt.xlabel('Category')
plt.ylabel('X')
plt.xticks(rotation=45)
plt.show()
```

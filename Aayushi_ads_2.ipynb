**ADS lab 2**

Dataset : Crop Production in India ([Link](https://www.kaggle.com/datasets/abhinand05/crop-production-in-india))

This dataset provides a huge amount of information on crop production in India ranging from several years. It's a state-wise Indian crop production dataset.

Features

**State_Name**: The name of the state or region where the crop was cultivated.
**District_Name**: The name of the district within the state where the crop was cultivated.
**Crop_Year**: The year in which the crop was cultivated.
**Season**: The season during which the crop was cultivated (e.g., Kharif, Rabi, Whole Year).
**Crop**: The name or type of crop that was cultivated.
**Area**: The land area (in hectares or acres) on which the crop was cultivated.
**Production**: The quantity of crop produced (in metric tons or other relevant units).

```python
missing_values = df.isnull().sum()
print(missing_values)
```

```python
# 1. Fill missing values with the maximum value in the column
df_max = df.fillna(df.max())

print("\nCleaned Dataset - Max Value:")
print(df_max.head())
```

```python
missing_values = df_max.isnull().sum()
print(missing_values)
```

```python
# 2. Fill missing values with the minimum value in the column
df_min = df.fillna(df.min())

print("\nCleaned Dataset - Min Value:")
print(df_min.head())
```

```python
missing_values = df_min.isnull().sum()
print(missing_values)
```

```python
drop_column = ["State_Name","District_Name","Season","Crop"]
df_drop = df.drop(columns = drop_column)
df_drop
```

```python
# 3. Fill missing values using k-means clustering
from sklearn.impute import KNNImputer
knn_imputer = KNNImputer(n_neighbors=2)
df_kmeans = pd.DataFrame(data=knn_imputer.fit_transform(df_drop), columns=df_drop.columns)
print("\nCleaned Dataset - K-means:")
print(df_kmeans.head())
```

```python
missing_values = df_kmeans.isnull().sum()
print(missing_values)
```

```python
# 4. Fill missing values with a fixed value
df_fixed_value = df.fillna(0)

print("\nCleaned Dataset - Fixed Value:")
print(df_fixed_value.head())
```

```python
missing_values = df_fixed_value.isnull().sum()
print(missing_values)
```

```python
# 5. Fill missing values with the most frequent value in the column
df_most_frequent = df.apply(lambda x: x.fillna(x.value_counts().index[0]))

print("\nCleaned Dataset - Most Frequent Value:")
print(df_most_frequent.head())
```

```python
missing_values = df_most_frequent.isnull().sum()
print(missing_values)
```

```python
# 6. Fill missing values using linear interpolation
df_interpolated = df.interpolate()

# Display cleaned datasets
print("\nCleaned Dataset - Interpolated:")
print(df_interpolated.head())
```

```python
missing_values = df_interpolated.isnull().sum()
print(missing_values)
```

```python
# Impute missing values in 'Production' column with median as there are outliers
median_production = df['Production'].median()
df['Production'].fillna(median_production, inplace=True)
df
```

```python
missing_values = df.isnull().sum()
print(missing_values)
```

```python
# Check for duplicate rows
duplicate_rows = df[df.duplicated()]

# Display duplicate rows, if any
if not duplicate_rows.empty:
    print("Duplicate Rows:")
    print(duplicate_rows)
else:
    print("No duplicate rows found.")
```

```python
# Standardize text data
df['State_Name'] = df['State_Name'].str.lower()
df['District_Name'] = df['District_Name'].str.lower()
df['Season'] = df['Season'].str.strip().str.lower()
df['Crop'] = df['Crop'].str.strip().str.lower()
```

```python
# Check consistency in categorical variables
print("Unique Seasons:", df['Season'].unique())
print("Unique Crops:", df['Crop'].unique())
```

```python

# 'Area' should be non-negative, and 'Production' should be non-negative and within a reasonable upper limit
reasonable_area_range = (0, float('inf'))  # Area should be non-negative
reasonable_production_range = (0, 100000)  # Production should be non-negative and less than 100,000 (example upper limit)

# Perform data validation for 'Area'
invalid_area_values = df[(df['Area'] < reasonable_area_range[0]) | (df['Area'] > reasonable_area_range[1])]
if not invalid_area_values.empty:
    print("Invalid 'Area' values found:")
    print(invalid_area_values)

# Perform data validation for 'Production'
invalid_production_values = df[(df['Production'] < reasonable_production_range[0]) | (df['Production'] > reasonable_production_range[1])]
if not invalid_production_values.empty:
    print("Invalid 'Production' values found:")
    print(invalid_production_values)
```

```python
import pandas as pd
from scipy import stats

#z-scores for 'Area' and 'Production' columns
z_scores_area = stats.zscore(df['Area'])
z_scores_production = stats.zscore(df['Production'])

# Define threshold for outlier detection (e.g., z-score > 3 or < -3)
threshold = 3

# Identify outliers based on threshold
outliers_area = df[abs(z_scores_area) > threshold]
outliers_production = df[abs(z_scores_production) > threshold]

# Remove outliers
cleaned_df = df[(abs(z_scores_area) <= threshold) & (abs(z_scores_production) <= threshold)]

# Display summary after outlier removal
print("Summary after outlier removal:")
print("Original dataset shape:", df.shape)
print("Cleaned dataset shape:", cleaned_df.shape)
```

```python
print(cleaned_df)
cleaned_df.to_csv('cleaned_data.csv', index=False)
```

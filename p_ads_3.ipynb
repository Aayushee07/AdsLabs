```python
import pandas as pd

df=pd.read_csv('/content/hotel_bookings.csv')
df
```

```python
import matplotlib.pyplot as plt

# Generate bubble graph
plt.figure(figsize=(10, 6))
plt.scatter(df['stays_in_weekend_nights'], df['stays_in_week_nights'], s=df['adults']*100, alpha=0.5)

# Set labels and title
plt.xlabel('Stays in Weekend Nights')
plt.ylabel('Stays in Week Nights')
plt.title('Bubble Graph of Stays')

plt.grid(True)
plt.show()

# The x-axis represents the number of stays in weekend nights (stays_in_weekend_nights).
# The y-axis represents the number of stays in weeknights (stays_in_week_nights).
# The size of the bubbles represents the number of adults (adults).
```

```python
import matplotlib.pyplot as plt

# Generate histogram
plt.figure(figsize=(10, 6))
plt.hist(df['adults'], bins=20, edgecolor='black', color='skyblue')

# Set labels and title
plt.xlabel('Number of Adults')
plt.ylabel('Frequency')
plt.title('Histogram of Number of Adults')

plt.grid(True)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Set Seaborn style
sns.set(style="whitegrid")

# Create countplot
plt.figure(figsize=(10, 6))
sns.countplot(x='market_segment', data=df, palette='pastel')

# Set labels and title
plt.xlabel('market_segment')
plt.ylabel('Count')
plt.title('Count of market_segment')

# Rotate x-axis labels for better readability
plt.xticks(rotation=45)

plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Set Seaborn style
sns.set(style="whitegrid")

# Create countplot with hue
plt.figure(figsize=(10, 6))
sns.countplot(x='customer_type', hue='market_segment', data=df, palette='pastel')

# Set labels and title
plt.xlabel('customer_type')
plt.ylabel('Count')
plt.title('Distribution of Hotel Types by Market Segment')

plt.show()
```

```python
import matplotlib.pyplot as plt

# Create scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(df['adults'], df['children'], color='skyblue', alpha=0.5)

# Set labels and title
plt.xlabel('Number of Adults')
plt.ylabel('Number of Children')
plt.title('Scatter Plot of Adults vs Children')

plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt

# Count the occurrences of each market segment
market_segment_counts = df['market_segment'].value_counts()

# Create pie chart
plt.figure(figsize=(8, 8))
plt.pie(market_segment_counts, labels=market_segment_counts.index, autopct='%1.1f%%', colors=['skyblue', 'lightgreen', 'lightcoral', 'orange'])
plt.title('Distribution of Market Segments')
plt.show()
```
